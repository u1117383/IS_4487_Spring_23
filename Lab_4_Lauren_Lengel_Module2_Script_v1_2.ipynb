## Clean up the data


```python
# taking out the outliers

# doing df and then [the column you want to reference... ex. house] 
# this will filter in... telling the data what to keep
  # ex. keep all houses over $0 & keep all incomes over $0 etc

# cleaning for zipcodes less than 5 digits
df_clean = df[(df['postal_code'] > 9999.00)]
df_clean.describe()
```

```python
# converting the Total_Amt into a number (not a string)
df_clean = df
df_clean["total_amount"] = df_clean["total_amount"].str.replace(",", "").str.replace("$", "").astype(float)
```

```python
# dropping the extra columns and extraneous info

df_clean = df[["city", "state_province", "customer_id",	"postal_code", "order_id", "order_date",	"product_id", "product_name", "quantity", "product_price", "total_amount", "order_month"]]
df_clean.info()
```

```python
# Checking for NULL values

df_clean.isnull().sum()     
```

```python
# Dropping the NULL values in the remaining rows that don't pertain to the analysis of their sale figures
# The city NULL values aren't very helpful so I'll drop them

drop_nulls_df = df.dropna(subset=["city"])
print(drop_nulls_df)
```
